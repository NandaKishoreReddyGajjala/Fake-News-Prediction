## Imports

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
import re
import joblib
import string
```

## Reading Data

## Data Preprocessing

```python
# Creating Class Column [1=True & 0=False]
true['class']=1
fake['class']=0
```

```python
# Concatenating both datasets
df = pd.concat([fake,true], axis=0)
```

```python
df.sample(5)
```

```python
# Dropping unnecessary columns
df = df.drop(["title", "subject", "date"], axis=1)
```

```python
df.sample(5)
```

```python
df.head()
```

```python
# Function to clean the text
def clean_text(text):
    text = text.lower()
    text = re.sub(r"\[.*?\]", "", text)
    text = re.sub(r"//W", "", text)
    text = re.sub(r"https?://\S+|www\.\S+", "", text)
    text = re.sub(r"<.*?>+", "", text)
    text = re.sub(r"[%s]" % re.escape(string.punctuation), "", text)
    text = re.sub(r"\n", "", text)
    text = re.sub(r"\w*\d\w*", "", text)
    return text
```

```python
# Applying the function to text column
df["text"] = df["text"].apply(clean_text)
```

## Data Modelling

```python
x = df["text"]
Y = df["class"]
```

```python
# Splitting Data into Train & Test data
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.2, random_state=42)
```

```python
#Initializing Vectorizer to convert text into numbers for the machine to understand
vectorizer = TfidfVectorizer()
xv_train = vectorizer.fit_transform(x_train)
xv_test = vectorizer.transform(x_test)
```

```python
# Model Initializing
model = LogisticRegression()
model.fit(xv_train, Y_train)
```

```python
# Prediction
prediction = model.predict(xv_test)
```

```python
# Calculating Accuracy
model.score(xv_test, Y_test)
```

```python
print(classification_report(Y_test,prediction))
```

```python
joblib.dump(vectorizer,"vectorizer.jb")
joblib.dump(model,"model.jb")
```
